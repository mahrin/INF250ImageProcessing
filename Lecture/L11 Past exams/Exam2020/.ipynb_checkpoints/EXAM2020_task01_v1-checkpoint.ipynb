# Exercise 1: Filters-Lab4
The image drone_blurr.tif shown in Figure 1 turned out a bit blurry. Use one or several of
the sharpening technques you have learned and see if you can make the image sharper. It is
ok to transform the image to an 8-bit image and work only on this.
Describe the technique(s) you use and show the results.

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.filters import median
from skimage.filters import gaussian
from skimage.morphology import disk
from skimage import data, img_as_float
from skimage.color import rgb2gray
from skimage import io

from skimage.restoration import denoise_nl_means
```

```python
filename = 'drone_blurr.tif'
drone_RGB = io.imread(filename)
plt.imshow(drone_RGB)
plt.title(label = "Original RGB image", fontsize=12)
plt.show()

drone_gray = rgb2gray(drone_RGB)
plt.imshow(drone_gray, cmap='gray')
plt.title(label = "Original GRAY image", fontsize=12)
plt.show()

gaussdrone = gaussian(drone_gray, sigma=10)
plt.imshow(gaussdrone,'gray')
plt.title(label = "Blurred image", fontsize=12)
plt.show()


# unsharpening mask to sharpen the image
amount = 3
usm_drone = drone_gray + amount*(drone_gray-gaussdrone)
plt.imshow(usm_drone, 'gray')
plt.title(label = "sharpened", fontsize=12)
plt.show()
```

```python
#string fault

#reading image
filename = 'string.jpg'
pv1 = io.imread(filename)


# display results 
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (16, 8))

# 0=red, 1=green, 2=blue
ax[0].imshow(pv1[:,:,0], cmap ='gray')
ax[0].set_title("red image PV1", fontsize = 14)

ax[1].imshow(pv1[:,:,1], cmap ='gray')
ax[1].set_title(r"green image PV1", fontsize = 14)

ax[2].imshow(pv1[:,:,2], cmap ='gray')
ax[2].set_title(r"blue image PV1", fontsize = 14)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

```python
filename = 'module.tif'
pv2 = io.imread(filename)

# display results 
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (16, 8))

# 0=red, 1=green, 2=blue
ax[0].imshow(pv2[:,:,0], cmap ='gray')
ax[0].set_title("red image pv2", fontsize = 14)

ax[1].imshow(pv2[:,:,1], cmap ='gray')
ax[1].set_title(r"green image pv2", fontsize = 14)

ax[2].imshow(pv2[:,:,2], cmap ='gray')
ax[2].set_title(r"blue image pv2", fontsize = 14)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

```python
#selecting the band with the most signal
pv1g = pv1[50:450,100:500,2] # CHose blue band

pv2g = pv2[100:400,100:500,2] # CHose blue band

# display results 
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (16, 8))

# 0=red, 1=green, 2=blue
ax[0].imshow(pv1g, cmap ='gray')
ax[0].set_title("red image pv2", fontsize = 14)

ax[1].imshow(pv2g, cmap ='gray')
ax[1].set_title(r"green image pv2", fontsize = 14)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

```python
from skimage import exposure

plt.figure()
plt.imshow(pv2g,'gray')
plt.title(label='Original image PV2g')
plt.show()

pv2eq = exposure.equalize_hist(pv2g)

plt.figure()
plt.imshow(pv2eq,'gray')
plt.title(label='Equalised contrasted image PV2g')
plt.show()


plt.hist(pv2g.ravel(),256, [0, 50], color='black')
plt.show()

plt.hist(pv2eq.ravel(),256, [0,50], color='black')
plt.show()
```

```python
radius = 15
selem = disk(radius)
local_otsu = rank.otsu(pv2g, selem)


threshold_global_otsu = threshold_otsu(pv2g)
global_otsu = pv2g >= threshold_global_otsu


threshold_global_otsu = threshold_otsu(pv1g)
global_otsu = pv1g >= threshold_global_otsu



# display results 
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (16, 8))


ax[0].imshow(pv2g>= local_otsu, cmap ='gray')
ax[0].set_title(label = "Local Otsu pv2", fontsize = 14)

ax[1].imshow(global_otsu, cmap ='gray')
ax[1].set_title("Global Otsu pv2", fontsize = 14)

ax[2].imshow(global_otsu, cmap ='gray')
ax[2].set_title(r"Global Otsu pv1", fontsize = 14)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

# Exercise 3
In this task you will look at another image of a solar panel. The image you downloaded called UVfluor.tif shows a fluorescence effect and has a rather low contrast.\
In the image we look for dark areas that are cracks in the solar panel. \
From the image UVfluor.tif we want to obtain images like the ones in Figure 3 in order to detect cracks in the solar panels.
- Describe how you obtain these and include the scripts you used.


```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.morphology import disk
from skimage.filters import threshold_otsu, rank
from skimage import exposure
from skimage.color import rgb2gray
```

```python
#reading image
filename = 'UVfluorim.tif'
from skimage import io
pv1 = io.imread(filename)

# display image
plt.figure()
plt.imshow(pv1)

#selecting the band with the most signal
pv1g = pv1[:,:,2]


# display image
plt.figure()
plt.imshow(pv1g,cmap='gray')


```

```python
# histogram

histogram = np.zeros(256)
shape = np.shape(pv1g)
for i in range(shape[0]):
    for j in range(shape[1]):
        pixval = int(pv1g[i,j])
        histogram[pixval] += 1
plt.figure()        
plt.plot(histogram)
plt.title(label='Histogram PV1g')
plt.show()


#histogram equalisation
# Cumulated histogram

cumhist = np.zeros(256)
cumhist[0] = histogram[0]
for i in range(255):
    cumhist[i+1] = cumhist[i] + histogram[i+1]

plt.figure()    
plt.plot(cumhist)
plt.title(label='Cumhist PV1g')
plt.show()

# histogram equalisation
pv2 = pv1g
K = 256
M = shape[0]
N = shape[1]
for i in range(shape[0]):
    for j in range(shape[1]):
        a = int(pv2[i,j])
        b = cumhist[a]*(K-1)/(M*N)
        pv2[i,j] = b
        


plt.figure()
plt.hist(pv2.ravel(), bins =256, range =[0,256])
plt.title(label='Hist equalisation PV1g')
plt.show()


plt.imshow(pv2,cmap='gray')
plt.title(label='Image after hist equalisation PV1g')
plt.show()
```

```python
# trehshold with Otsu
radius = 35
selem = disk(radius)
local_otsu = rank.otsu(pv1g, selem)
plt.figure()
plt.imshow(pv1g>= local_otsu, cmap='gray')
plt.title(label='Local otsu PV1g')
plt.show()

threshold_global_otsu = threshold_otsu(pv2)
global_otsu = pv2 >= threshold_global_otsu


plt.figure()
plt.imshow(global_otsu, cmap='gray')
plt.title(label='gloal otsu PV1g')
plt.show()
```
