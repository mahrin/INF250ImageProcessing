```python
# Solution v1: Ingun
filename = 'drone_blurr.tif'
drone_RGB = io.imread(filename)
plt.imshow(drone_RGB)
plt.title(label = "Original RGB image", fontsize=12)
plt.show()

drone_gray = rgb2gray(drone_RGB)
plt.imshow(drone_gray, cmap='gray')
plt.title(label = "Original GRAY image", fontsize=12)
plt.show()

gaussdrone = gaussian(drone_gray, sigma=10)
plt.imshow(gaussdrone,'gray')
plt.title(label = "Blurred image", fontsize=12)
plt.show()


# unsharpening mask to sharpen the image
amount = 3
usm_drone = drone_gray + amount*(drone_gray-gaussdrone)
plt.imshow(usm_drone, 'gray')
plt.title(label = "sharpened", fontsize=12)
plt.show()
```

# Task1: Solution v3 (With drone and 3 sharpening): Own version

```python
def sharpen(image, sharpmask):
    """Performs an image sharpening using Laplace filter or unsharpen mask (USM)
    1 = Laplace
    2 = USM
    3 = high pass
    
    Returns: sharpened image
    """
    amount = 3 # parameter--hypertuning, weight of sharpening
    if sharpmask == 1:
        filtered = laplace(image) 
        sharpened = image - amount*filtered
    elif sharpmask == 2:  
        gaussasian_blur = gaussian(image, sigma=10) 
        sharpened = image + amount*(image - gaussasian_blur) 
    elif sharpmask == 3:  
        gaussasian_blur = gaussian(image, sigma=10) 
        sharpened = image - gaussasian_blur  
    elif sharpmask == 4:  
        gaussasian_blur = gaussian(image, sigma=5) 
        sharpened = image + amount*(image - gaussasian_blur) 
    elif sharpmask == 5:  
        gaussasian_blur = gaussian(image, sigma=5) 
        sharpened = image - gaussasian_blur  
    elif sharpmask == 6:  
        gaussasian_blur = gaussian(image, sigma=15) 
        sharpened = image + amount*(image - gaussasian_blur) 
    elif sharpmask == 7:  
        gaussasian_blur = gaussian(image, sigma=15) 
        sharpened = image - gaussasian_blur  
    else:
        print("wrong mask")
    return sharpened
```

```python
filename = 'drone_blurr.tif'
drone_RGB = io.imread(filename)
drone_gray = rgb2gray(drone_RGB)

#cropping the drone
drone_gray= drone_gray[1215:1440, 1285:1580]   #[Row Y, COL, X]
# plt.imshow(drone_gray, cmap='gray')
# plt.title(label = "Original GRAY image", fontsize=12)
# plt.show()

# calling all the sharpening methods
laplace_sharp = sharpen(drone_gray[:,:], 1)
usm_sharp = sharpen(drone_gray[:,:], 2)
high_pass_sharp = sharpen(drone_gray[:,:], 3)
usm_sharp2 = sharpen(drone_gray[:,:], 4)
high_pass_sharp2 = sharpen(drone_gray[:,:], 5)
usm_sharp3 = sharpen(drone_gray[:,:], 6)
high_pass_sharp3 = sharpen(drone_gray[:,:], 7)

#Viewing the images
show_plot_sharp()
```

```python
def show_plot_sharp():

    fig, ax = plt.subplots(nrows = 3, ncols = 3, figsize = (12, 10))

    ax[0, 0].imshow(drone_gray, cmap ='gray')
    ax[0, 0].set_title('original', fontsize = 10)
    
    ax[0, 1].imshow(laplace_sharp, cmap ='gray')
    ax[0, 1].set_title(r'laplace_sharp', fontsize = 10)
    
    ax[0, 2].imshow(drone_gray, cmap ='gray')
    ax[0, 2].set_title('original', fontsize = 10)
    

    
    ax[1, 0].imshow(usm_sharp, cmap ='gray')
    ax[1, 0].set_title(r'usm_sharp sigma 5', fontsize = 10)
    
    ax[1, 1].imshow(usm_sharp2, cmap ='gray')
    ax[1, 1].set_title(r'usm_sharp sigma 10', fontsize = 10)

    ax[1, 2].imshow(usm_sharp3, cmap ='gray')
    ax[1, 2].set_title(r'usm_sharp sigma 15', fontsize = 10)    
    
    
    
    ax[2,0].imshow(high_pass_sharp, cmap ='gray')
    ax[2,0].set_title(r'high_pass_sharp sigma 5', fontsize = 10)
    
    ax[2,1].imshow(high_pass_sharp2, cmap ='gray')
    ax[2,1].set_title(r'high_pass_sharp sigma 10', fontsize = 10)
 
    ax[2,2].imshow(high_pass_sharp3, cmap ='gray')
    ax[2,2].set_title(r'high_pass_sharp sigma 15', fontsize = 10)
    
    for row in ax:
        for col in row:
            col.axis('off')

    fig.tight_layout()
    plt.show()
```

# Task3: V1: Solution using manual codes: Ingun

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray

from skimage.morphology import disk

from skimage.filters import rank
from skimage.filters import try_all_threshold
from skimage.filters import threshold_otsu, threshold_minimum, threshold_triangle, threshold_mean


from skimage import exposure
```

```python
#reading image
filename = 'UVfluorim.tif'
from skimage import io
pv1 = io.imread(filename)

# display image
plt.figure()
plt.imshow(pv1)
plt.title(label='Original image')
plt.show()

#selecting the band with the most signal
pv1g = pv1[:,:,2]

# display image
plt.figure()
plt.imshow(pv1g,cmap='gray')
plt.title(label='Blue band image')
plt.show()
```

```python
#histogram equalisation

# histogram
histogram = np.zeros(256)
shape = np.shape(pv1g)
for i in range(shape[0]):
    for j in range(shape[1]):
        pixval = int(pv1g[i,j])
        histogram[pixval] += 1
plt.figure()        
plt.plot(histogram)
plt.title(label='Histogram PV1g')
plt.show()



# Cumulated histogram
cumhist = np.zeros(256)
cumhist[0] = histogram[0]
for i in range(255):
    cumhist[i+1] = cumhist[i] + histogram[i+1]

plt.figure()    
plt.plot(cumhist)
plt.title(label='Cumhist PV1g')
plt.show()

# histogram equalisation
pv2 = pv1g[:, :]
K = 256
M = shape[0]
N = shape[1]
for i in range(shape[0]):
    for j in range(shape[1]):
        a = int(pv2[i,j])
        b = cumhist[a]*(K-1)/(M*N)
        pv2[i,j] = b
        


plt.figure()
plt.hist(pv2.ravel(), bins =256, range =[0,256])
plt.title(label='Hist equalisation of equalised image')
plt.show()


plt.imshow(pv2,cmap='gray')
plt.title(label='Image after hist equalisation')
plt.show()
```

```python
# trehshold with Otsu
local_otsu = rank.otsu(pv1g, disk(radius= 35))

plt.figure()
plt.imshow(pv1g>= local_otsu, cmap='gray')## DId not understand why??
plt.title(label='Local thresholding otsu PV1g')
plt.show()

# This image has very low signals, so the local thresholding will not work. 
#It would amplify the noise present in the image

threshold_global_otsu = threshold_otsu(pv2)
global_otsu = pv2 >= threshold_global_otsu

plt.figure()
plt.imshow(global_otsu, cmap='gray')
plt.title(label='global thresholding otsu PV1g')
plt.show()
```
