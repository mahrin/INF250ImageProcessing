```python
def get_image(name):
    """ Using the skimage.io import the image with the file name.
    This function returns the gray image as numpy array """
    
    #Getting the image  
    filename = '../Images/'+name
    
    #https://scikit-image.org/docs/stable/api/skimage.io.html#skimage.io.imread
    image = io.imread(filename, as_gray = True) #changing the image to grayscale when reading, so RGBA to GRAY scale
#     print("Orginal shape", image.shape)

    return image
```

## Different edge operator filters

```python
def edge_operator(image, operator):
    """Returns the reusult from one of the edge operators, prewitt, sobel,
    canny OR laplace
    
    Parameters:
    -----------
    image : np.ndarray
        Image to detect blobs in. If this image is a colour image then 
        the last dimension will be the colour value (as RGB values).
    operator : numeric
    1 = sobel filter
    2 = prewitt filter
    3 = canny filter
    4 = laplace filter

    Returns:
    --------
    filtered : np.ndarray(np.uint)
    result image from the edge operator
    """
    #Reading the image
    image = get_image(image)
    
    # Setup
    shape = np.shape(image)
    if len(shape) == 3:#making sure it is a color/RGB image
        image = image.mean(axis=2)
    elif len(shape) > 3:
        raise ValueError('Must be at 2D image')
    
    # performing the filters
    filtered = None
    if operator == 1:
        filtered = filters.sobel(image)
    elif operator == 2:
        filtered = filters.prewitt(image) 
    elif operator == 3:
        filtered = feature.canny(image, sigma=3)        
    elif operator == 4:
        filtered = filters.laplace(image)
    else:
        return "you have given wrong operator"
        
    return filtered
```

# display results for all filters WITHOUT vmin and vmax

```python
# display results for all filters WITHOUT vmin and vmax
def show_results_for_all():
    fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 8))

    ax[0,0].imshow(sobel, cmap='hot')
    ax[0,0].set_title('sobel', fontsize = 10)

    ax[0, 1].imshow(prewitt, cmap='hot')
    ax[0, 1].set_title(r'prewitt', fontsize = 10)

    ax[1, 0].imshow(canny, cmap='hot')
    ax[1, 0].set_title(r'canny', fontsize = 10)

    ax[1, 1].imshow(laplace, cmap='hot')
    ax[1, 1].set_title(r'laplace', fontsize = 10)

    for row in ax:
        for col in row:
            col.axis('off')

    fig.tight_layout()
    plt.show()
```

```python
#Calling the different filters
sobel = edge_operator("AthenIR.png", 1)
prewitt = edge_operator("AthenIR.png", 2)
canny = edge_operator("AthenIR.png", 3)
laplace = edge_operator("AthenIR.png", 4)
```

```python
# Viewing all the filter results
show_results_for_all()
```

## Different image sharpening masks

# Sharpening
 1. Shapening using Laplace edge detector
 2. Shapening using High pass filter which works utilising low-pass-filter
 3. Shapening using Unsharpend Mask (USM)


one application of a Laplacian operator is to restore fine detail to an image which has been smoothed to remove noise. 
(The median operator is often used to remove noise in an image.)

https://www.oreilly.com/library/view/hands-on-image-processing/9781789343731/04fdee54-7183-4598-a9ec-546bb67d2622.xhtml

https://stackoverflow.com/questions/39619222/laplacian-sharpening-grey-image-as-result



```python
def sharpen(image, sharpmask):
    """Performs an image sharpening using Laplace filter or unsharpen mask (USM)
    1 = Laplace
    2 = USM
    
    Returns: sharpened image
    """
    amount = 2 # parameter--hypertuning, weight of sharpening
    if sharpmask == 1:
        filtered = filters.laplace(image) 
        sharpened = image - amount*filtered
    elif sharpmask == 2:  
        gaussasian_blur = filters.gaussian(image, sigma=5) 
        sharpened = image + amount*(image - gaussasian_blur) 
    else:
        print("wrong mask")
    return sharpened
```

```python
image = get_image("AthenIR.png")
laplace_sharp = sharpen(image, 1)
usm_sharp = sharpen(image, 2)
```

```python
#finding max and min pixel value of a image
smallest = np.amin(laplace_sharp[:,2])
biggest = np.amax(laplace_sharp [:,2])
print("smallest Laplace", smallest )
print("biggest Laplace", biggest)

print("=======================")

smallest = np.amin(usm_sharp[:,2])
biggest = np.amax(usm_sharp [:,2])
print("smallest USM", smallest )
print("biggest USM", biggest)

print("=======================")
```
