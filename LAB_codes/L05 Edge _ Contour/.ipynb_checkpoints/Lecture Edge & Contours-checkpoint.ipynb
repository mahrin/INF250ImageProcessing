```python
import numpy as np
import matplotlib.pyplot as plt 

from skimage import data, img_as_float 
from skimage.data import astronaut 
from skimage.color import rgb2gray

from skimage import filters
from skimage.filters import gaussian, laplace 
```

```python
#importing and making the image gray
astro = rgb2gray(img_as_float(data.astronaut())) 
astro = astro[30:180,150:300] #cropping the head 
plt.imshow(astro,'gray')
```

# Edge detectors
1. Prewitt
2. Sobel
3. Canny
4. Laplace

```python
#Psobel and prewitt filter
astro_sobel = filters.sobel(astro) #can get vertical and horizontal images
astro_sobel_h = filters.sobel_h(astro) 
astro_sobel_v = filters.sobel_v(astro) 
astro_prewitt = filters.prewitt(astro) #can get ONLY the total image

# display results
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(14, 6))

ax[0].imshow(astro_prewitt, cmap='gray')
ax[0].set_title(r'astro_prewitt', fontsize=10)

ax[1].imshow(astro_sobel, cmap='gray')
ax[1].set_title('astro_sobel', fontsize=10)

ax[2].imshow(astro_sobel_h, cmap='gray')
ax[2].set_title(r'astro_sobel_h', fontsize=10)

ax[3].imshow(astro_sobel_v, cmap='gray')
ax[3].set_title(r'astro_sobel_v', fontsize=10)



for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

```python
#Canny filter
from skimage import feature
edges1 = feature.canny(astro, sigma=1) 
# plt.imshow(edges1,'gray')
# plt.show()

edges2 = feature.canny(astro, sigma=2) 
# plt.imshow(edges2,'gray')
# plt.show()

edges3 = feature.canny(astro, sigma=3) 
edges4 = feature.canny(astro, sigma=5) 
edges5 = feature.canny(astro, sigma=2.5) 


# display results
fig, ax = plt.subplots(nrows=1, ncols=5, figsize=(16, 10))

ax[0].imshow(edges1, cmap='gray')
ax[0].set_title('Canny with sigma=1', fontsize=10)

ax[1].imshow(edges2, cmap='gray')
ax[1].set_title(r'Canny with sigma=2', fontsize=10)

ax[2].imshow(edges5, cmap='gray')
ax[2].set_title(r'Canny with sigma=2.5', fontsize=10)

ax[3].imshow(edges3, cmap='gray')
ax[3].set_title(r'Canny with sigma=3', fontsize=10)


ax[4].imshow(edges4, cmap='gray')
ax[4].set_title(r'Canny with sigma=5', fontsize=10)

plt.show()
```

```python
# Laplace filter
astro_lap = laplace(image = astro) # zick zack of white & black
plt.imshow(astro_lap,'gray',vmin=0.1, vmax=0.2) 
plt.title('astro_lap edge', fontsize=10)
plt.show()
```

# Sharpening
 1. Shapening using Laplace edge detector
 2. Shapening using High pass filter
 3. Shapening using Unsharpend Mask (USM)

```python
# laplace filter

astro_lap = laplace(image = astro) # zick zack of white & black
# plt.imshow(astro_lap,'gray',vmin=0.1, vmax=0.2) 

astro_sharp = astro - 2*astro_lap #finding the sharpened image, with weight of shrap=2
# plt. imshow(astro_sharp, 'gray', vmin=0.4, vmax=0.9)


# display results
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))

ax[0].imshow(astro_lap,'gray',vmin=0.1, vmax=0.2) 
ax[0].set_title('astro_lap edge', fontsize=10)

ax[1].imshow(astro_sharp, 'gray', vmin=0.4, vmax=0.9)
ax[1].set_title(r'astro_edge sharp', fontsize=10)
plt.show()
```

```python
# # high pass filter
# from skimage.filters import gaussian 
# plt.imshow(astro, 'gray')

# gaussastro = gaussian(astro, sigma=5) 
# plt.imshow(gaussastro,'gray')

# astro_high = astro-(gaussastro) 
# plt.imshow(astro_high,'gray', vmin=0., vmax=0.2)
```

```python
# high pass filter

gaussastro1 = gaussian(astro, sigma=2) 
astro_high1 = astro-(gaussastro1) 

gaussastro2 = gaussian(astro, sigma=3) 
astro_high2 = astro-(gaussastro2) 

gaussastro3 = gaussian(astro, sigma=5) 
astro_high3 = astro-(gaussastro3) 


# USM Mask # unsharpening mask to sharpen the image
amount = 2 # parameter--hypertuning, weight of sharpening
# FORMULA usm =  I + a.M = original + sharpness_weight*(original - blur_image)
usm_astro1 = astro + amount*(astro - gaussastro1) 
usm_astro2 = astro + amount*(astro - gaussastro2) 
usm_astro3 = astro + amount*(astro - gaussastro3) 


amount = 5 # parameter--hypertuning, weight of sharpening
# FORMULA usm =  I + a.M = original + sharpness_weight*(original - blur_image)
usm_astro4 = astro + amount*(astro - gaussastro1) 
usm_astro5 = astro + amount*(astro - gaussastro2) 
usm_astro6 = astro + amount*(astro - gaussastro3) 



# display results
fig, ax = plt.subplots(nrows=3, ncols=5, figsize=(20, 12))

ax[0, 0].imshow(astro, 'gray')
ax[0, 0].set_title('Original image', fontsize=10)

ax[0, 1].imshow(gaussastro1,'gray')
ax[0, 1].set_title(r'gaussasian with sigma = 2', fontsize=10)

ax[0, 2].imshow(astro_high1,'gray', vmin=0., vmax=0.2)
ax[0, 2].set_title(r'High filter after subtraction (sigma=2)', fontsize=10)

ax[0, 3].imshow(usm_astro1,'gray', vmin=0., vmax=0.2)
ax[0, 3].set_title(r'USM sharpening (sigma=2, amoun=2)', fontsize=10)

ax[0, 4].imshow(usm_astro4,'gray', vmin=0., vmax=0.2)
ax[0, 4].set_title(r'USM sharpening (sigma=2, amoun=5)', fontsize=10)




ax[1, 0].imshow(astro, 'gray')
ax[1, 0].set_title('Original image', fontsize=10)

ax[1,1].imshow(gaussastro2,'gray')
ax[1,1].set_title(r'gaussasian with sigma = 3', fontsize=10)

ax[1, 2].imshow(astro_high2,'gray', vmin=0., vmax=0.2)
ax[1, 2].set_title(r'High filter after subtraction (sigma=3)', fontsize=10)

ax[1, 3].imshow(usm_astro2,'gray', vmin=0., vmax=0.2)
ax[1, 3].set_title(r'USM sharpening (sigma=3, amoun=2)', fontsize=10)

ax[1, 4].imshow(usm_astro5,'gray', vmin=0., vmax=0.2)
ax[1, 4].set_title(r'USM sharpening (sigma=3, amoun=5)', fontsize=10)




ax[2, 0].imshow(astro, 'gray')
ax[2, 0].set_title('Original image', fontsize=10)

ax[2, 1].imshow(gaussastro3,'gray')
ax[2, 1].set_title(r'gaussasian with sigma = 5', fontsize=10)

ax[2, 2].imshow(astro_high3,'gray', vmin=0., vmax=0.2)
ax[2, 2].set_title(r'High filter after subtraction (sigma=5)', fontsize=10)

ax[2, 3].imshow(usm_astro3,'gray', vmin=0., vmax=0.2)
ax[2, 3].set_title(r'USM sharpening (sigma=5, amoun=2)', fontsize=10)

ax[2, 4].imshow(usm_astro5,'gray', vmin=0., vmax=0.2)
ax[2, 4].set_title(r'USM sharpening (sigma=5, amoun=5)', fontsize=10)


plt.show()
```

# Canny filter [Code from skimage website]
#Source: https://scikit-image.org/docs/dev/auto_examples/edges/plot_canny.html \
#Documentation: https://scikit-image.org/docs/stable/api/skimage.feature.html#skimage.feature.canny


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage.util import random_noise
from skimage import feature


# Generate noisy image of a square
image = np.zeros((128, 128), dtype=float) # image of 0s--black
image[32:-32, 32:-32] = 1 # making 32x32 block white
image = ndi.rotate(image, 15, mode='constant') #rotating the image
image = ndi.gaussian_filter(image, 4) #blurring the image
image = random_noise(image, mode='speckle', mean=0.1) #added noise
plt.imshow(image, cmap='gray')
```

```python
# Compute the Canny filter for two values of sigma
edges1 = feature.canny(image) #by default sigma = 1
edges2 = feature.canny(image, sigma=3)

# display results
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 3))

ax[0].imshow(image, cmap='gray')
ax[0].set_title('noisy image', fontsize=20)

ax[1].imshow(edges1, cmap='gray')
ax[1].set_title(r'Canny filter, $\sigma=1$', fontsize=20)

ax[2].imshow(edges2, cmap='gray')
ax[2].set_title(r'Canny filter, $\sigma=3$', fontsize=20)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

# Laplace sharpening using OpenCV

```python
def show_image(image, label ="Image:", size=(8,3)):
    """This function shows/plots the given image."""
    plt.subplots(1,1,figsize=size)
    plt.title(label = label, fontsize=12)
    plt.imshow(image, cmap='gray') 
    plt.show()
```

```python
from skimage.filters import laplace
from cv2 import imread as imread
im = rgb2gray(imread('../images/airfield.tif'))
im1 = np.clip(laplace(im) + im, 0, 1)

show_image(im, 'original image', (6,6))
show_image(im1, 'sharpened image', (6,6))
```
