```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage import filters
from skimage import data, img_as_float 
from skimage.data import astronaut 
from skimage.color import rgb2gray
```

```python
#importing and making the image gray
astro = rgb2gray(img_as_float(data.astronaut())) 
astro = astro[30:180,150:300] #cropping the head 
plt.imshow(astro,'gray')
```

```python
# #Psobel and prewitt filter
# astro_sobel = filters.sobel(astro) 
# plt.imshow(astro_sobel,'gray')
# plt.show()

# astro_sobel_h = filters.sobel_h(astro) 
# plt.imshow(astro_sobel_h,'gray')
# plt.show()

# astro_sobel_v = filters.sobel_v(astro) 
# plt.imshow(astro_sobel_v,'gray')
# plt.show()

# astro_prewitt = filters.prewitt(astro) 
# plt.imshow(astro_prewitt,'gray')
# plt.show()
```

```python
#Psobel and prewitt filter
astro_sobel = filters.sobel(astro) #can get vertical and horizontal images
astro_sobel_h = filters.sobel_h(astro) 
astro_sobel_v = filters.sobel_v(astro) 
astro_prewitt = filters.prewitt(astro) #can get ONLY the total image

# display results
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(10, 4))

ax[0].imshow(astro_sobel, cmap='gray')
ax[0].set_title('astro_sobel', fontsize=10)

ax[1].imshow(astro_sobel_h, cmap='gray')
ax[1].set_title(r'astro_sobel_h', fontsize=10)

ax[2].imshow(astro_sobel_v, cmap='gray')
ax[2].set_title(r'astro_sobel_v', fontsize=10)

ax[3].imshow(astro_prewitt, cmap='gray')
ax[3].set_title(r'astro_prewitt', fontsize=10)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

```python
#Canny filter
from skimage import feature
edges1 = feature.canny(astro, sigma=1) 
# plt.imshow(edges1,'gray')
# plt.show()

edges2 = feature.canny(astro, sigma=3) 
# plt.imshow(edges2,'gray')
# plt.show()


# display results
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))

ax[0].imshow(edges1, cmap='gray')
ax[0].set_title('Canny with sigma=1', fontsize=10)

ax[1].imshow(edges2, cmap='gray')
ax[1].set_title(r'Canny with sigma=2', fontsize=10)
```

```python
# laplace filter
from skimage.filters import laplace
astro_lap = laplace(astro) # zick zack of white & black
# plt.imshow(astro_lap,'gray',vmin=0.1, vmax=0.2) 

astro_sharp = astro - 2*astro_lap #finding the sharpened image, with weight of shrap=2
# plt. imshow(astro_sharp, 'gray', vmin=0.4, vmax=0.9)


# display results
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))

ax[0].imshow(astro_lap,'gray',vmin=0.1, vmax=0.2) 
ax[0].set_title('astro_lap edge', fontsize=10)

ax[1].imshow(astro_sharp, 'gray', vmin=0.4, vmax=0.9)
ax[1].set_title(r'astro_edge sharp', fontsize=10)
```

```python
# high pass filter
from skimage.filters import gaussian 

gaussastro = gaussian(astro, sigma=5) 
astro_high = astro-(gaussastro) 

# display results
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 3))

ax[0].imshow(astro, 'gray')
ax[0].set_title('Original image', fontsize=10)

ax[1].imshow(gaussastro,'gray')
ax[1].set_title(r'gaussasian with sigma = 5', fontsize=10)

ax[2].imshow(astro_high,'gray', vmin=0., vmax=0.2)
ax[2].set_title(r'High filter after subtraction', fontsize=10)
```

```python
# unsharpening mask to sharpen the image
# amount = 2 # parameter--hypertuning
# usm_astro = astro + amount*(astro-gaussastro) 
# plt.imshow(usm_astro, 'gray', vmin=0.2, vmax=0.9)
# plt.imshow(astro,'gray', vmin=0.2, vmax=0.9)
# plt.hist(astro.ravel(),256,[0,1], color='black') 
# plt.show()

# plt.hist(gaussastro.ravel(),256,[0,1], color='black') 
# plt.show()
```

```python
# unsharpening mask to sharpen the image
amount = 2 # parameter--hypertuning, weight of sharpening
usm_astro = astro + amount*(astro - gaussastro) 
# FORMULA usm =  I + a.M = original + sharpness_weight*(original - blur_image)

# display results
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))

ax[0].imshow(astro,'gray', vmin=0.2, vmax=0.9)
ax[0].set_title('Original image', fontsize=10)

ax[1].imshow(usm_astro, 'gray', vmin=0.2, vmax=0.9)
ax[1].set_title(r'Usm result', fontsize=10)
```

```python
#compraing the histogram
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(14, 5))

ax[0].hist(astro.ravel(),256,[0,1], color='black') 
ax[0].set_title(r'Histogram of original', fontsize=10)

ax[1].hist(gaussastro.ravel(),256,[0,1], color='black') 
ax[1].set_title(r'Histogram of Gaussian', fontsize=10)

ax[2].hist(usm_astro.ravel(),256,[0,1], color='black') 
ax[2].set_title(r'Histogram of usm_astro', fontsize=10)
```

```python
#Canny filter
# Source: https://scikit-image.org/docs/dev/auto_examples/edges/plot_canny.html
#Documentation: https://scikit-image.org/docs/stable/api/skimage.feature.html#skimage.feature.canny

import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage.util import random_noise
from skimage import feature


# Generate noisy image of a square
image = np.zeros((128, 128), dtype=float) # image of 0s--black
image[32:-32, 32:-32] = 1 # making 32x32 block white
image = ndi.rotate(image, 15, mode='constant') #rotating the image
image = ndi.gaussian_filter(image, 4) #blurring the image
image = random_noise(image, mode='speckle', mean=0.1) #added noise
plt.imshow(image, cmap='gray')
```

```python
# Compute the Canny filter for two values of sigma
edges1 = feature.canny(image) #by default sigma = 1
edges2 = feature.canny(image, sigma=3)

# display results
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 3))

ax[0].imshow(image, cmap='gray')
ax[0].set_title('noisy image', fontsize=20)

ax[1].imshow(edges1, cmap='gray')
ax[1].set_title(r'Canny filter, $\sigma=1$', fontsize=20)

ax[2].imshow(edges2, cmap='gray')
ax[2].set_title(r'Canny filter, $\sigma=3$', fontsize=20)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

```python
def show_image(image, label ="Image:", size=(8,3)):
    """This function shows/plots the given image."""
    plt.subplots(1,1,figsize=size)
    plt.title(label = label, fontsize=12)
    plt.imshow(image, cmap='gray') 
    plt.show()
```

```python
from skimage.filters import laplace
from cv2 import imread as imread
im = rgb2gray(imread('../images/airfield.tif'))
im1 = np.clip(laplace(im) + im, 0, 1)

show_image(im, 'original image', (10,6))
show_image(im1, 'sharpened image', (10,6))
```
