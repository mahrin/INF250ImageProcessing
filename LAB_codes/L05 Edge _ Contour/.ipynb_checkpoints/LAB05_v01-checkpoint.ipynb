```python
# importing all the necessary modules
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from skimage import filters #sobel,  prewitt & laplace
from skimage import feature# canny

#for colors
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
```

```python
def get_image(name):
    """ Using the skimage.io import the image with the file name.
    This function returns the gray image as numpy array """
    
    #Getting the image  
    filename = '../Images/'+name
    image = io.imread(filename)
#     print("Orginal shape", image.shape)
    image = image.mean(axis = 2)
#     print("After mean shape", image.shape)
#     print(image[2][2])

    return image
```

```python
def show_image(image, label ="Image:", size=(8,3)):
    """This function shows/plots the given image."""
    plt.subplots(1,1,figsize=size)
    plt.title(label = label, fontsize=12)
    plt.imshow(image, cmap='gray') 
    plt.show()
```

```python

def edge_operator(image, operator):
    """Returns the reusult from one of the edge operators, prewitt, sobel,
    canny eller laplace
    
    Parameters:
    -----------
    image : np.ndarray
        Image to detect blobs in. If this image is a colour image then 
        the last dimension will be the colour value (as RGB values).
    operator : numeric
    1 = sobel filter
    2 = prewitt filter
    3 = canny filter
    4 = laplace filter

    Returns:
    --------
    filtered : np.ndarray(np.uint)
    result image from the edge operator
    """
    #Reading the image
    image = get_image(image)
    
    # performing the filters
    filtered = None
    if operator == 1:
        filtered = filters.sobel(image)
    elif operator == 2:
        filtered = filters.prewitt(image) 
    elif operator == 3:
        filtered = feature.canny(image, sigma = 15)        
    elif operator == 4:
        filtered = filters.laplace(image) 
        
    return filtered
```

```python
sobel = edge_operator("AthenIR.png", 1)
prewitt = edge_operator("AthenIR.png", 2)
canny = edge_operator("AthenIR.png", 3)
laplace = edge_operator("AthenIR.png", 4)
show_results_for_all()
```

```python
# display results
def show_results_for_all():
    fig, ax = plt.subplots(nrows = 1, ncols = 4, figsize = (12, 6))

    ax[0].imshow(sobel, cmap='gray')
    ax[0].set_title('sobel', fontsize = 10)

    ax[1].imshow(prewitt, cmap='gray')
    ax[1].set_title(r'prewitt', fontsize = 10)

    ax[2].imshow(canny, cmap='gray')
    ax[2].set_title(r'canny', fontsize = 10)

    ax[3].imshow(laplace, cmap='gray')
    ax[3].set_title(r'laplace', fontsize = 10)

    for a in ax:
        a.axis('off')

    fig.tight_layout()
    plt.show()
```

one application of a Laplacian operator is to restore fine detail to an image which has been smoothed to remove noise. 
(The median operator is often used to remove noise in an image.)

https://www.oreilly.com/library/view/hands-on-image-processing/9781789343731/04fdee54-7183-4598-a9ec-546bb67d2622.xhtml

https://stackoverflow.com/questions/39619222/laplacian-sharpening-grey-image-as-result

```python
def sharpen(image, sharpmask):
    """Performs an image sharpening using Laplace filter or unsharpen mask (USM)
    1 = Laplace
    2 = USM
    
    Returns: sharpened image
    """
    amount = 1 # parameter--hypertuning, weight of sharpening
    if sharpmask == 1:
        filtered = filters.laplace(image) 
        sharpened = image - amount*filtered
    elif sharpmask == 2:  
        gaussasian_blur = filters.gaussian(image, sigma = 5) 
        sharpened = image + 0.8*(image - gaussasian_blur) 
    else:
        print("wrong mask")
    return sharpened
```

```python
image = get_image("AthenIR.png")
show_image(image)
show_image(sharpen(image, 2), "usm_sharp")
```
