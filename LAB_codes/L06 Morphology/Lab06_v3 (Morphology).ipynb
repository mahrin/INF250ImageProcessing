##### What is the function of  the min/max filter in relation to   erode and dilate?

**Erode(shrink)**
- Erosion on the original image makes the holes huge, but removes **unnecessary small ojects**
- Erosion on on the dilated image, make it back to the original size


**dilate(grow)** 
- dilate on the image multiple time, **removes holes** 

=============================================================================


##### What is the effect of open and close ?


1. Open

    - Performs an erosion operation, followed by dilation. 
    - This smoothes objects and removes isolated pixels/ **removes noises**.

2. Close
    - Performs a dilation operation, followed by erosion. 
    - This smoothes objects and **fills in small holes**.

```python
from skimage.morphology import skeletonize, dilation, erosion, opening, closing, square, disk
from skimage import data
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import numpy as np
from skimage import io
from skimage.filters import threshold_mean#for binarising
from skimage.filters import median 

#for max & min filter
from skimage.morphology import disk, ball
from skimage.filters.rank import mean, maximum, minimum 
from skimage import img_as_ubyte as byte_convo #Convert an image to unsigned byte format, with values in [0, 255].


from skimage import feature# canny
```

```python
def show_image(image, label ="Image:", size=(6,2)):
    """This function shows/plots the given image."""
    plt.subplots(1,1,figsize=size)
    plt.title(label = label, fontsize=12)
    plt.imshow(image, cmap='gray') 
    plt.show()
```

```python
filename = 'rhino_d.tif'
rhino = io.imread(filename)
```

```python
rhino1 = byte_convo(rhino) # oRiginally was 0 and 255
print(rhino1)
show_image(rhino1, "", (6,4))
```

```python
rhino2 = rhino.astype('uint8')# Changed the (0, 255) to (0, 1) 
print(rhino2)
show_image(rhino2, "", (6, 4))
```

### dilate (using square(3))
- dilate(bigger) on the image multiple time, removes holes 

```python
filename = 'rhino_d.tif'
rhino = io.imread(filename)

print("Removing holes with dialation of white parts/foreground")
rhinodil1 = dilation(rhino,square(3))
rhinodil2 = dilation(rhinodil1,square(3))
rhinodil3 = dilation(rhinodil2,square(3))


#Plotting the image
fig, ax = plt.subplots(nrows = 1, ncols = 4, figsize = (14, 8))

ax[0].imshow(rhino, cmap ='gray')
ax[0].set_title("Original image", fontsize = 10)

ax[1].imshow(rhinodil1, cmap ='gray')
ax[1].set_title("dilation 1 removing holes", fontsize = 10)

ax[2].imshow(rhinodil2, cmap ='gray')
ax[2].set_title("dilation 2 removing holes", fontsize = 10)

ax[3].imshow(rhinodil3, cmap ='gray')
ax[3].set_title("dilation 3 removing holes", fontsize = 10)



for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

### erode  (using square(3))
1. Erode(shrink) on the original image makes the holes huge, but removes unnecessary small ojects
2. Erode(shrink) on the dilated image, make the back to original size

```python
print("Removing noises with erosion")

rhinoerode = erosion(rhino,square(3)) # With the original image
rhinoerode1 = erosion(rhinoerode,square(3))
#2 time was enough for going back to original size, 4 makes it too thin/small
rhinoerode2 = erosion(rhinoerode1,square(3))
rhinoerode3 = erosion(rhinoerode2,square(3))


#Plotting the image
fig, ax = plt.subplots(nrows = 1, ncols = 4, figsize = (14, 8))

ax[0].imshow(rhino, cmap ='gray')
ax[0].set_title("Original image", fontsize = 10)

ax[1].imshow(rhinoerode, cmap ='gray')
ax[1].set_title("erosion 1", fontsize = 10)

ax[2].imshow(rhinoerode2, cmap ='gray')
ax[2].set_title("erosion 2", fontsize = 10)

ax[3].imshow(rhinoerode3, cmap ='gray')
ax[3].set_title("erosion 3", fontsize = 10)


for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

### dilate followed by erode: Close

```python
#dilate followed by erode: Close
filename = 'rhino_d.tif'
rhino = io.imread(filename)

# Dilation, then erosion
rhinodil1 = dilation(rhino, square(3))
rhinoerode = erosion(rhinodil1, square(3)) # With the original image

#close on original image
rhinoclose = closing(rhino,square(3))



#Plotting the image
fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (12, 6))

fig.suptitle("mimicking close:", fontsize = 14)

ax[0, 0].imshow(rhino, cmap ='gray')
ax[0, 0].set_title("Original image", fontsize = 10)

ax[0, 1].imshow(rhinodil1, cmap ='gray')
ax[0, 1].set_title("dilation", fontsize = 10)

ax[0, 2].imshow(rhinoerode, cmap ='gray')
ax[0, 2].set_title("erosion on dilated", fontsize = 10)



ax[1, 0].imshow(rhino, cmap ='gray')
ax[1, 0].set_title("Original image", fontsize = 10)

ax[1, 1].imshow(rhino, cmap ='gray')
ax[1, 1].set_title("Original image", fontsize = 10)

ax[1, 2].imshow(rhinoclose, cmap ='gray')
ax[1, 2].set_title("Close", fontsize = 10)


for row in ax:
    for col in row:
        col.axis('off')

fig.tight_layout()
plt.show()
```

### erode followed by dilate:Open

```python
#erode followed by dilate:Open

filename = 'rhino_d.tif'
rhino = io.imread(filename)

#Dilation on the eroded image
rhinoerode = erosion(rhino, square(4)) # With the original image
rhinodil1 = dilation(rhinoerode, square(4))

#Open on the original image
rhinoopen = opening(rhino,square(4))


#Plotting the image
fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (13, 6))

fig.suptitle("mimicking  Open:", fontsize = 14)

ax[0, 0].imshow(rhino, cmap ='gray')
ax[0, 0].set_title("Original image", fontsize = 10)

ax[0, 1].imshow(rhinoerode, cmap ='gray')
ax[0, 1].set_title("Erosion", fontsize = 10)

ax[0, 2].imshow(rhinodil1, cmap ='gray')
ax[0, 2].set_title("Dilation on Eroded", fontsize = 10)


ax[1, 0].imshow(rhino, cmap ='gray')
ax[1, 0].set_title("Original image", fontsize = 10)

ax[1, 1].imshow(rhino, cmap ='gray')
ax[1, 1].set_title("Original image", fontsize = 10)

ax[1, 2].imshow(rhinoopen , cmap ='gray')
ax[1, 2].set_title("Open:erode-dilate", fontsize = 10)


for row in ax:
    for col in row:
        col.axis('off')

fig.tight_layout()
plt.show()
```

### Maximum Filter

- max (disc radius 1 og 3)

Applying the maximum filter removes the **negative outlier noise** present in a Digital Image.
When a maximum filter is applied, the darker objects present in the image are eroded. This way maximum filter is called an **erosion filter**. With respect to the lighter pixels, some call this as a dilation filter.

- https://pythontic.com/image-processing/pillow/maximum%20filter

```python
#Disk1 & DIsk3 maximum

filename = 'rhino_d.tif'
rhino = io.imread(filename)

#disk1
img_maximum_disk1 = maximum( rhino.astype('uint8'), footprint = disk(1) )

#disk3
img_maximum_disk3 = maximum( rhino.astype('uint8'), footprint = disk(3) )


#Plotting the image
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (14, 8))
fig.suptitle("Max filtere:", fontsize = 14)

ax[0].imshow(rhino, cmap ='gray')
ax[0].set_title('original', fontsize = 10)

ax[1].imshow(img_maximum_disk1, cmap ='gray')
ax[1].set_title("maximum filter with disk 1", fontsize = 10)

ax[2].imshow(img_maximum_disk3, cmap ='gray')
ax[2].set_title("maximum filter with disk 3", fontsize = 10)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

### Minimum Filter

- min (disc radius 1 og 3)

**Minimum filter** is also called as a **dilation filter**. When minimum filter is applied the **object boundaries** present in an image are **extended**.



```python
#Disk3 Minimum

filename = 'rhino_d.tif'
rhino = io.imread(filename)


img_minimum_disk1 = minimum( rhino.astype('uint8'), footprint = disk(1) )
img_minimum_disk3 = minimum( rhino.astype('uint8'), footprint = disk(3) )


#Plotting the image
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (14, 8))
fig.suptitle("Min filtere:", fontsize = 14)

ax[0].imshow(rhino, cmap ='gray')
ax[0].set_title('original', fontsize = 10)

ax[1].imshow(img_minimum_disk1, cmap ='gray')
ax[1].set_title("minimum filter with disk 1", fontsize = 10)

ax[2].imshow(img_minimum_disk3, cmap ='gray')
ax[2].set_title("minimum filter with disk 3", fontsize = 10)


for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

## 2) Contours 

Work on the image rhino_detail.tif and see if you can trace the contours of the object as shown below.

### Solution-v1 (own version)

```python
filename = 'rhino_detail.tif'
image = io.imread(filename)
show_image(image, "", (6,4))
```

```python
#Dilating for filling hole of the image
def fill_holes(image):
    for i in range(3):
        filled_dialated_image = dilation(image, square(3))
    filled_dialated_image = np.invert(filled_dialated_image)
    return filled_dialated_image
```

```python
filled_dialated_image = fill_holes(image)
show_image(filled_dialated_image, "", (6,4))
```

```python
def get_eroded_inverted(image):
    #Eroding the image
    eroded_image = image
    for i in range(3):
        eroded_image = erosion(eroded_image, square(3))
 
    # Inverting the image
    inverted_eroded_image = np.invert(eroded_image[:, :])
    
    return inverted_eroded_image
```

```python
inverted_eroded_image = get_eroded_inverted(filled_dialated_image)
show_image(inverted_eroded_image, "", (6,4))
```

```python
# # CHECKING THE SIZES
# show_image(image, "Original Binary image")
# show_image(filled_dialated_image, "filled_dialated_image")
# show_image(inverted_eroded_image, "inverted_eroded_image")

# print("Original Binary image shape", image.shape)
# print("filled_dialated image shape", filled_dialated_image.shape)
# print("Inverted eroded image shape", inverted_eroded_image.shape)
```

```python
#Getting the contour
contour = filled_dialated_image.astype(int) + inverted_eroded_image.astype(int)

# Making the contour thinner
for i in range(2):
    contour = erosion(contour, square(2))
        
inverted_contour = np.invert(contour)

show_image(inverted_contour, "", (6,4))
```

```python
#Plotting the image
fig, ax = plt.subplots(nrows = 1, ncols = 4, figsize = (14, 8))

ax[0].imshow(image, cmap ='gray')
ax[0].set_title("Original image", fontsize = 10)

ax[1].imshow(filled_dialated_image, cmap ='gray')
ax[1].set_title("filled_dialated_image", fontsize = 10)

ax[2].imshow(inverted_eroded_image, cmap ='gray')
ax[2].set_title("inverted_eroded_image", fontsize = 10)

ax[3].imshow(inverted_contour, cmap ='gray')
ax[3].set_title("inverted_contour", fontsize = 10)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```

You will work on the image fingerprint.tif shown below. The goal is to improve the image so that other algorithms can perform fingerprint recognition. Use different filters to improve the image as much as you can.

#### Solution-v1(own)

```python
#Importing the image
filename = 'fingerprint.tif'
fingerprint_image = io.imread(filename)
# show_image(image, "Original image", (10,4))
```

```python
#Denoising image using erosion--shrinking
denoised = fingerprint_image
for i in range(2):
    denoised = erosion(denoised, square(2))

#Returning to original size_v1 : using disk1  
resized1 = denoised[:, :] 
for i in range(2):
    resized1 = dilation(resized1, disk(1))
    
#Returning to original size_v2  : Using square2
resized2 = denoised[:, :]
for i in range(2):
    resized2 = dilation(resized2, square(2))
    

#Plotting the image
fig, ax = plt.subplots(nrows = 1, ncols = 4, figsize = (15, 7))

ax[0].imshow(fingerprint_image, cmap ='gray')
ax[0].set_title("Original image", fontsize = 10)

ax[1].imshow(denoised, cmap ='gray')
ax[1].set_title("denoised (shrinked) with erosion Sqr2", fontsize = 10)

ax[2].imshow(resized1, cmap ='gray')
ax[2].set_title("resized denoised image Disk-1", fontsize = 10)

ax[3].imshow(resized2, cmap ='gray')
ax[3].set_title("resized denoised image SQUARE-2", fontsize = 10)


for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```
