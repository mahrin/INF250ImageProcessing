# Task1: import and show image

```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

#Getting the image  airfield.tif
filename = '../Images/airfield.tif'
#filename = '../Images/ireland.tif'
ireland = io.imread(filename)
plt.title(label = "Gray iamge of airfield", fontsize=12)
plt.imshow(ireland,cmap='gray')
```

# Task2: BASIC HISTOGRAM WITH 256 BINS

```python
histogram = np.zeros(256)
shape = np.shape(ireland)
for i in range(shape[0]):
    for j in range(shape[1]):
        pixval = int(ireland[i,j])
        histogram[pixval] += 1
plt.figure()       
plt.title(label="histogram of airfield", fontsize=12)
plt.plot(histogram)
```

# Task3: increasing contrast with 50 %

```python
# increasing contrast with 50 %
plt.imshow(ireland,cmap='gray') 
shape = np.shape(ireland)
for i in range(shape[0]):
    for j in range(shape[1]):
        ireland[i,j] = int(ireland[i,j]*1.5+0.5)
        if ireland[i,j] > 255:
            ireland[i,j] = 255
plt.imshow(ireland,cmap='gray')   
```

# Task4: histogram equalisation (with original image)



```python

# histogram equalisation with the original histogram
M = shape[0] #Row of image
N = shape[1] # Col of image

K = 256

for i in range(shape[0]):#Row
    for j in range(shape[1]):#Col
        a = int(ireland[i,j])
        #print("K", K)
        b = histogram[a]*(K-1)/(M*N)  # here, k is 256 
        ireland[i,j] = b
        
plt.title(label="histogram equalisation on original airfield", fontsize=12)
plt.imshow(ireland,'gray')
plt.figure()
```

```python
#plotting the histogram after the histogram equalisation
plt.subplots(1,1,figsize=(10,6))
plt.hist(ireland.ravel(), bins = 256, range =[0,256])
plt.show()
```

## Type2:  1st we need to have the Cumulated histogram

```python
#Getting the image  airfield.tif
filename = '../Images/airfield.tif'
ireland = io.imread(filename)


# making the orginal histogram
histogram = np.zeros(256)
shape = np.shape(ireland)
for i in range(shape[0]):
    for j in range(shape[1]):
        pixval = int(ireland[i,j])
        histogram[pixval] += 1
plt.figure()  

plt.title(label="histogram of airfield", fontsize=12)
plt.plot(histogram)


# Cumulated histogram
cumhist = np.zeros(256)
cumhist[0] = histogram[0]
for i in range(255):
    cumhist[i+1] = cumhist[i] + histogram[i+1]

plt.figure()   
plt.title(label="Cumulated histogram of airfield", fontsize=12)
plt.plot(cumhist)
```

# histogram equalisation with the Cumulated histogram

```python
# histogram equalisation with the Cumulated histogram
M = shape[0] #Row of image
N = shape[1] # Col of image

K = 256
for i in range(shape[0]):#Row
    for j in range(shape[1]):#Col
        a = int(ireland[i,j])
        #print("K", K)
        b = cumhist[a]*(K-1)/(M*N)  # here, k is 256 
        ireland[i,j] = b
        
plt.subplots(1,1,figsize=(8,4))        
plt.imshow(ireland,'gray')
plt.title(label="histogram equalisation with Cumulated histogram", fontsize=12)
plt.figure()


```

```python
plt.subplots(1,1,figsize=(25,10))
plt.hist(ireland.ravel(),256,[0,256])
plt.title(label="histogram after histogram equalisation with Cumulated histogram", fontsize=12)
plt.show()
```

```python
plt.subplots(1,1,figsize=(25, 10))
plt.hist(ireland.ravel(),256,[0,256])
plt.show()
```

# Task6:  Otsu threshold on the original image with skimage

```python
from skimage.filters import threshold_otsu

#Getting the image  
filename = '../Images/airfield.tif'

ireland = io.imread(filename)


thresh = threshold_otsu(ireland)
binary = ireland > thresh

plt.imshow(binary, cmap=plt.cm.gray)
plt.title(label = "Binary Otsu threshold image of airfield", fontsize=12)


plt.show()
```

# Task4 with skimage

```python
#TRY1
from skimage import exposure

#Getting the image  
filename = '../Images/airfield.tif'
ireland = io.imread(filename)

#executing the Histogram Equalization
camera_equalized = exposure.equalize_hist(ireland)

# showing the iamge after Histogram Equalization
plt.imshow(camera_equalized, cmap=plt.cm.gray)
plt.title(label = "Histogram Equalization with skimage airfield", fontsize=12)
plt.show()
```
