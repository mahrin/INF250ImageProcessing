````python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

from skimage.morphology import skeletonize, dilation, opening, square, disk
from skimage.morphology import erosion, closing
from skimage import data
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import numpy as np
from skimage import io


from skimage.filters import threshold_mean#for binarising
from skimage.filters import try_all_threshold
from skimage.filters import threshold_otsu, threshold_minimum, threshold_triangle, threshold_mean


from skimage.morphology import skeletonize, dilation, opening, square, disk
from skimage.morphology import erosion, closing
from skimage import data
from skimage.filters import threshold_mean#for binarising


from scipy.ndimage import binary_fill_holes# for filling the holes

# For watershed
from scipy import ndimage as ndi#for distance
from skimage.feature import peak_local_max 
from skimage.segmentation import watershed 
from skimage import measure #measure objects
````

````python
def get_image(name):
    """ Using the skimage.io import the image with the file name.
    This function returns the gray image as numpy array """
    
    #Getting the image  
    #filename = '../Images/'+name
    filename = name
    image = io.imread(filename, as_gray = True) 
    
    return image
````

````python
def show_image(image, label ="Image:", size=(6,2)):
    """This function shows/plots the given image."""
    plt.subplots(1,1,figsize=size)
    plt.title(label = label, fontsize=12)
    plt.imshow(image, cmap='gray') 
    plt.show()
````

````python
image = get_image("IMG_2754_nonstop_alltogether.JPG")
# show_image(image, "Original Binary image")

#cropping the head
image = image[420:3350, 650:4980]   #[Row, COL]
plt.imshow(image, 'gray')
# show_image(image, "cropped image")
````

````python
fig, ax = try_all_threshold(image, figsize=(15, 13), verbose=False)
plt.show()
````

````python
#binarising using minimum filter
thresh = threshold_minimum(image)
binary_minimum = image > thresh
show_image(binary_minimum, "Minimum threshold image", (10,5))
````

````python
#binarising using otsu filter
thresh = threshold_otsu(image)
binary_otsu = image > thresh
show_image(binary_otsu, "Otsu threshold image", (10,5))
````

````python
#binarising using mean filter
thresh = threshold_mean(image)
binary_mean = image > thresh
show_image(binary_mean, "mean threshold image", (15,8))
````

````python
#binarising using triangle filter
thresh = threshold_triangle(image)
binary_triangle = image > thresh
show_image(binary_triangle, "triangle threshold image", (15,8))
````

````python
# import SimpleITK as sitk
# from scipy import ndimage
# brain_mask_raw = binary_triangle
# brain_mask_array = sitk.GetArrayFromImage(brain_mask_raw)
# brain_mask_array_filled = binary_fill_holes(brain_mask_array, structure=np.ones((3,3,3))).astype(int)
````

````python
inverted_binary_triangle = np.invert(binary_triangle)
filled = binary_fill_holes(inverted_binary_triangle).astype(int)
# inverted_filled = np.invert(filled)
````

````python
show_image(filled, "filled triangle threshold image", (10,5))
# show_image(inverted_filled, "inverted filled triangle threshold image", (10,5))
````

````python
#watershed

#Distance map
distance = ndi.distance_transform_edt(filled) 
plt.imshow(distance, 'gray')
plt.show()
````

````python
#watershed using distance map
local_maxi = peak_local_max(distance, min_distance =100, indices=False, footprint=np.ones((3,3)), labels=filled)
markers = ndi.label(local_maxi)[0]
labels = watershed(-distance,markers, mask=filled)
plt.figure() 
plt.imshow(labels)
plt.show()
````

````python
# #measure objects
# labels2 = measure.label(filled)

# properties = measure. regionprops(labels)
# for prop in properties:
#     print(prop.perimeter, prop.area, prop.centroid, prop.eccentricity)
````

````python
# From the website of ImageJ we know, 
# ```    
# Circ. (circularity) = (4*area)/perimeter^2 
# ```
import math

#measure objects
labels2 = measure.label(filled)

properties = measure.regionprops(labels2)

data={}
# object_no=1
min_val =0
max_val = 0
for object_no, prop in enumerate(properties):
    #print(prop.perimeter, prop.area, prop.centroid)
    if prop.area>60:
        circularity = ((4*math.pi*prop.area)/(prop.perimeter**2))
        data[object_no]= [prop.area,circularity]
    if min_val> prop.area:
        min_val= prop.area
    elif max_val<prop.area:
        max_val= prop.area
````

````python
print(data)
print(len(data))
print("max_val", max_val)
print("min_val", min_val)
````
